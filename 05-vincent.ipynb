# More Brains

Segmented my brain into 3 distinct tissue classes ("Other", Gray & White matter)

```python
from nibabel import load
data = load('data/brain_stripped_seg.nii.gz').get_data()
hist, edges = np.histogram(data, bins=4)
bar = vincent.Bar(list(hist))
bar.axis_titles(x='Tissue Class', y='Voxel Count')
```

# Just some slicing & dicing...

```python
all_vals = list(np.unique(data))
columns = {0: 'Air',
           1: 'Other',
           2: 'Gray',
           3: 'White'}
cols = ['Other', 'Gray', 'White']

def get_axial(slice):
    return data[:, :, slice]

def get_sagittal(slice):
    return data[slice, :, :]

def get_coronal(slice):
    return data[:, slice, :]

def df_from_view(view='axial'):
    lu = {'axial': (2, get_axial),
          'coronal': (1, get_coronal),
          'sagittal': (0, get_sagittal)}
    dim_index, get_data = lu[view]
    slice_data = []
    for slice_index in range(0, data.shape[dim_index]):
        sd = {}
        sliced = get_data(slice_index).flatten()
        for value in all_vals:
            sd[value] = len(sliced[sliced==value])
        slice_data.append(sd)
    return pd.DataFrame(data=slice_data).rename(columns=columns)[cols]

sag_df = df_from_view(view='sagittal')
cor_df = df_from_view(view='coronal')
ax_df = df_from_view()
```

# Left to Right

```python
sag = vincent.StackedArea(sag_df)
sag.legend(title="Tissue Class")
sag.axis_titles(x="Slice (<----Left)", y="Voxel Count")
sag.colors(brew='Set2')
```

# Bottom to Top

```python
ax = vincent.StackedArea(ax_df)
ax.legend(title="Tissue Class")
ax.axis_titles(x="Slice (<----Neck)", y="Voxel Count")
ax.colors(brew='Set3')
```

# Front to Back

```python
cor = vincent.StackedArea(cor_df)
cor.legend(title="Tissue Class")
cor.axis_titles(x="Slice (<----Front)", y="Voxel Count")
cor.colors(brew='Spectral')
```
