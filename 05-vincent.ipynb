# Stacked{Bar,Area}

```python
df = pd.DataFrame(data={'asc': range(0, 100, 2),
                        'des': range(100, 0, -2)})
sb = vincent.StackedBar(df)
sb.legend(title='ranges')
sb.display()
```

```python
sa = vincent.StackedArea(df)
sa.axis_titles(x='Index', y='Value')
# ColorBrewer
sa.colors(brew='Set3')
```

```python
data = load('data/brain_stripped_seg.nii.gz').get_data()
hist, edges = np.histogram(data, bins=4)
```

```python
vincent.Bar(list(hist))
```

```python
all_vals = list(np.unique(data))

slice_data = []
for slice_index in range(0, data.shape[0]):
    sd = {}
    sliced = data[slice_index, :, :].flatten()
    for value, count in zip(all_vals, np.bincount(sliced)):
        sd[value] = count
    slice_data.append(sd)
df = pd.DataFrame(data=slice_data)
df
```

```python
sa = vincent.StackedArea(df)
sa.legend(title="Tissue Class")
```
